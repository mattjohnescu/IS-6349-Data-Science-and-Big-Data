## Task 8b

## Task 9

```python
# change the transparency of the lower half of the image, and display itself 
# Making copy of image to avoid modifying the original
arch_png_opacity = arch_png.copy()

# Starting row index for the lower half
start_row_index = arch_png_opacity.shape[0] // 2

# Set the opacity (alpha) value to 0.4 for the lower half of the image
arch_png_opacity[start_row_index:, :, 3] = 0.4

# Display the modified image
print("PNG Image with Lower Half Transparency Set to 0.4:")
display_image(arch_png_opacity)
```

## Task 10

```python
# display a negative of the png image
# Copy of the image to avoid modifying original
arch_png_negative = arch_png.copy()

# Reverse the colors for each channel (RGB) separately
arch_png_negative[:, :, 0] = 1.0 - arch_png_negative[:, :, 0]  # Red channel
arch_png_negative[:, :, 1] = 1.0 - arch_png_negative[:, :, 1]  # Green channel
arch_png_negative[:, :, 2] = 1.0 - arch_png_negative[:, :, 2]  # Blue channel

# Display the negative image
print("Negative PNG Image:")
display_image(arch_png_negative)


#BONUS 1 line Answer
arch_png_negative_1 = arch_png.copy()

# Reverse the colors for all channels (RGB) in one line, keeping opacity the same
arch_png_negative_1[:, :, :3] = 1.0 - arch_png_negative_1[:, :, :3]

# Display the negative image
print("Negative PNG Image:")
display_image(arch_png_negative_1)
```

## Task 11

```python
# obtain and display a cropping from the png image
# Start and end indices for cropping
row_start, row_end = 3500, 3800
col_start, col_end = 1800, 2200

# Crop to the specified area
arch_png_cropped = arch_png[row_start:row_end, col_start:col_end, :]

#Trying to zoom in on the piece of wood at the bottom of the frame
# Display the cropped image
print("Cropped PNG Image:")
display_image(arch_png_cropped)
```

##  Task 12

```python
# Display png image converted to grayscale, print its shape to verify
# Define the weights for RGB channels
weights = np.array([0.21, 0.72, 0.07])

# Convert the image to grayscale using weighted average
arch_png_gray = np.dot(arch_png[:, :, :3], weights)

# Print the shape of the resulting grayscale image to verify it is 2D
print(f"Shape of the grayscale PNG image: {arch_png_gray.shape}")

# Display the grayscale image
plt.figure(figsize=(8, 12))
plt.imshow(arch_png_gray, cmap='gray')
plt.xlabel('columns')
plt.ylabel('rows')
plt.colorbar()
plt.show()
```

## Task 13

```python
# Display two  gamma correctied images - light and dark
# Image 1
# Define gamma value
gamma = 1/5

# Apply Gamma correction to the RGB channels
arch_png_gamma_light = arch_png.copy()
arch_png_gamma_light[:, :, :3] = np.power(arch_png_gamma_light[:, :, :3], gamma)

# Gamma corrected image (lighter)
print("PNG Image with Gamma Correction (lighter):")
display_image(arch_png_gamma_light)

# Image 2
# Gamma value
gamma = 5

# Gamma correction to the RGB channels
arch_png_gamma_dark = arch_png.copy()
arch_png_gamma_dark[:, :, :3] = np.power(arch_png_gamma_dark[:, :, :3], gamma)

# Gamma corrected image (darker)
print("PNG Image with Gamma Correction (darker):")
display_image(arch_png_gamma_dark)
```

## Task 14

```python
# obtain the img in red, green and blue scale and horizontally stack/ join them 
# Create redscale image
img_png_Red = arch_png.copy()
img_png_Red[:, :, [1, 2]] = 0

# Create greenscale image
img_png_Green = arch_png.copy()
img_png_Green[:, :, [0, 2]] = 0

# Create bluescale image 
img_png_Blue = arch_png.copy()
img_png_Blue[:, :, [0, 1]] = 0

# Horizontally stack the images
horizontally_stacked_img = np.hstack((img_png_Red, img_png_Green, img_png_Blue))

# Display the horizontally stacked image
print("Horizontally Stacked Image (Redscale, Greenscale, Bluescale):")
display_image(horizontally_stacked_img)
```
