## Task 11. Create new computed columns

### Answer 11.i

```python
# Compute the number of years
num_years = 2019 - 2015

# Compute the CAGR for each state
popDF['CAGR_2015_2019'] = ((popDF['y2019'] / popDF['y2015']) ** (1 / num_years)) - 1

# Round the CAGR values to 4 decimal places
popDF['CAGR_2015_2019'] = popDF['CAGR_2015_2019'].round(4)

# Show the first three rows of only the new column
popDF[['CAGR_2015_2019']].head(3)
```

### Answer 11.ii

```python
# Define a function to compute the coefficient of variation
def coeff_of_variation(row):
    return row.std() / row.mean()

# Select the columns for the years 2010-2019
columns = ['y2010', 'y2011', 'y2012', 'y2013', 'y2014', 'y2015', 'y2016', 'y2017', 'y2018', 'y2019']

# Compute the coefficient of variation for each state
popDF['coeffofvar'] = popDF[columns].apply(coeff_of_variation, axis=1)

# Display the first 5 values of the new column
popDF[['coeffofvar']].head(5)
```

### Answer 11.iii

```python
# Compute the mean of the population in 2019 grouped by region
mean2019popbyReg = popDF.groupby('Region')['y2019'].transform('mean')

# Add the new column to the DataFrame
popDF['mean2019popbyReg'] = mean2019popbyReg

# Print the first 5 rows of the new column
popDF[['mean2019popbyReg']].head()
```
